## Plots of Raw Means for SNAP Participation

#### I show the raw means of SNAP participation using Matplotlib

### Importing Libraries

```python
import pandas as pd

df = pd.read_stata("df_analysis.dta")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Define states that  expanded Medicaid 

```python
df["medic"] = 0

expansion_states = [
    4, 5, 6, 8, 9, 10, 11, 15, 17, 19, 21, 24, 25,
    26, 27, 32, 33, 34, 35, 36, 38, 39, 41, 44,
    50, 53, 54, 18, 2, 30, 22
]

df.loc[df["statefip"].isin(expansion_states), "medic"] = 1
```

### Plotting SNAP Participation (for all individuals in the data)

```python

# Keep years 2008â€“2020

df_plot = df[(df["year"] >= 2008) & (df["year"] <= 2020)].copy()


# Collapse: mean SNAP1 by year
snap_year = (
    df_plot
    .groupby("year", as_index=False)["SNAP1"]
    .mean()
)


# Plot
plt.figure()
plt.plot(
    snap_year["year"],
    snap_year["SNAP1"],
    marker="o",
    linewidth=2
)

# Vertical line at Medicaid expansion
plt.axvline(2014, linestyle="--")

# Axes formatting
plt.xlim(2008, 2020)
plt.xticks(range(2008, 2021))
plt.xlabel("Year")
plt.ylabel("Mean SNAP Participation")
plt.title("Raw SNAP Participation Over Time (All Individuals)")

# Clean white background (Stata-like)
plt.gca().set_facecolor("white")
plt.gcf().set_facecolor("white")

plt.show()
```

#### Ploting SNAP participation (for medicaid-eligible individuals)

```python

# Keep medicaid-eligible

df_plot = df[
    (df["medicaid_eligible"] == 1)
].copy()


# Collapse: mean SNAP1 by year
snap_year = (
    df_plot
    .groupby("year", as_index=False)["SNAP1"]
    .mean()
)

plt.figure()
plt.plot(
    snap_year["year"],
    snap_year["SNAP1"],
    marker="o",
    linewidth=2
)

# Medicaid expansion line
plt.axvline(2014, linestyle="--")

# Axes formatting
plt.xlim(2008, 2020)
plt.xticks(range(2008, 2021))
plt.xlabel("Year")
plt.ylabel("Mean SNAP Participation")
plt.title("Raw SNAP Participation (Medicaid-eligible individuals)")

# White background (Stata style)
plt.gca().set_facecolor("white")
plt.gcf().set_facecolor("white")

plt.show()
```

#### Plotting SNAP Participation for medicaid-eligible individuals (by state expansion status)

```python
# restriction
df_plot = df[
    (df["medicaid_eligible"] == 1) &
    (df["sample1"] == 1)
].copy()

# Collapse: mean SNAP1 by year and expansion status

snap_year = (
    df_plot
    .groupby(["year", "medic"], as_index=False)["SNAP1"]
    .mean()
)

# Split data
snap_expand = snap_year[snap_year["medic"] == 1]
snap_nonexpand = snap_year[snap_year["medic"] == 0]

# Plot

plt.figure()

plt.plot(
    snap_expand["year"],
    snap_expand["SNAP1"],
    linewidth=2,
    label="Medicaid expansion states"
)

plt.plot(
    snap_nonexpand["year"],
    snap_nonexpand["SNAP1"],
    linestyle="--",
    linewidth=2,
    label="Non-expansion states"
)

# Medicaid expansion line
plt.axvline(2014, linestyle="--")

# Axes formatting
plt.xticks(range(2008, 2021))
plt.xlabel("Year")
plt.ylabel("Mean SNAP Participation")
plt.title("SNAP Participation Over Time")

# Legend
plt.legend(fontsize="small")

# White background (Stata-style)
plt.gca().set_facecolor("white")
plt.gcf().set_facecolor("white")

plt.show()
```

```python
## Restriction

df_plot = df[
    (df["medicaid_eligible"] == 1) &
    (df["sample1"] == 1) &
    (df["year"].between(2008, 2020))
].copy()


```

#### Plotting SNAP Participation for medicaid-eligible individuals (by state expansion status and gender)

```python
fig, ax = plt.subplots()

##Collaspe

collapsed = (
    df_plot
    .groupby(["year", "medic", "female"], as_index=False)
    .agg(SNAP1=("SNAP1", "mean"))
)

# Expansion states
ax.plot(
    collapsed.query("medic == 1 & female == 1")["year"],
    collapsed.query("medic == 1 & female == 1")["SNAP1"],
    color="navy",
    linestyle="-",
    linewidth=2,
    label="Expansion â€“ Female",
)

ax.plot(
    collapsed.query("medic == 1 & female == 0")["year"],
    collapsed.query("medic == 1 & female == 0")["SNAP1"],
    color="navy",
    linestyle="--",
    linewidth=2,
    label="Expansion â€“ Male",
)

# Non-expansion states
ax.plot(
    collapsed.query("medic == 0 & female == 1")["year"],
    collapsed.query("medic == 0 & female == 1")["SNAP1"],
    color="maroon",
    linestyle="-",
    linewidth=2,
    label="Non-expansion â€“ Female",
)

ax.plot(
    collapsed.query("medic == 0 & female == 0")["year"],
    collapsed.query("medic == 0 & female == 0")["SNAP1"],
    color="maroon",
    linestyle="--",
    linewidth=2,
    label="Non-expansion â€“ Male",
)

# Policy line
ax.axvline(2014, linestyle="--", color="gray")

# Labels and title
ax.set_xlabel("Year")
ax.set_ylabel("Mean SNAP Participation")
ax.set_title("SNAP Participation (Medicaid Expansion and Gender)")

ax.set_xticks(range(2008, 2021))
ax.legend(fontsize="small")

plt.show()
```

#### Plotting SNAP Participation for medicaid-eligible individuals (by state expansion status and race) 

```python

# nonwhite indicator (matches Stata)
df_plot["nonwhite"] = (df_plot["white"] == 0).astype(int)

# Collapse: mean SNAP1 by year Ã— medic Ã— nonwhite
collapsed = (
    df_plot
    .groupby(["year", "medic", "nonwhite"], as_index=False)
    .agg(SNAP1=("SNAP1", "mean"))
)

# Plot
fig, ax = plt.subplots()

# Expansion states
ax.plot(
    collapsed.query("medic == 1 & nonwhite == 0")["year"],
    collapsed.query("medic == 1 & nonwhite == 0")["SNAP1"],
    color="navy", linestyle="-", linewidth=2,
    label="Expansion â€“ White",
)

ax.plot(
    collapsed.query("medic == 1 & nonwhite == 1")["year"],
    collapsed.query("medic == 1 & nonwhite == 1")["SNAP1"],
    color="navy", linestyle="--", linewidth=2,
    label="Expansion â€“ Non-White",
)

# Non-expansion states
ax.plot(
    collapsed.query("medic == 0 & nonwhite == 0")["year"],
    collapsed.query("medic == 0 & nonwhite == 0")["SNAP1"],
    color="maroon", linestyle="-", linewidth=2,
    label="Non-expansion â€“ White",
)

ax.plot(
    collapsed.query("medic == 0 & nonwhite == 1")["year"],
    collapsed.query("medic == 0 & nonwhite == 1")["SNAP1"],
    color="maroon", linestyle="--", linewidth=2,
    label="Non-expansion â€“ Non-White",
)

# Formatting (mirror Stata)
ax.axvline(2014, linestyle="--", color="gray")
ax.set_xlim(2008, 2020)
ax.set_xticks(range(2008, 2021))
ax.set_xlabel("Year")
ax.set_ylabel("Mean SNAP Participation")
ax.set_title("SNAP Participation (Medicaid Expansion and Race)")

ax.legend(
    fontsize="small",
    loc="upper left",
    frameon=False,
)

plt.show()
```

#### Plotting SNAP Participation for medicaid-eligible individuals (by state expansion status and High-school completion)
